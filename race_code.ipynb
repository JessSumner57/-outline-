```python
# read data from all sheets and store them in a separate dataframe

def ethnic_loop(file_name):
    '''
    This function reads data from each sheet and
    put them in the same dataframe.
    
    '''
    
    # excel sheet 2020 to 2012
    result = pd.DataFrame(columns = ['code', 'borough', 'white', 'asian', 'black', 'mixed_other', 'year'])
    for i in range(1, 10):
       
        X = pd.read_excel(file_name, sheet_name = i, usecols = 'A:F', skipfooter = 15, skiprows = 4,
                          header = None, names = ['code', 'borough', 'white', 'asian', 'black', 'mixed_other'])
        X['year'] = 2021 - int(f'{i}')
        
        result = pd.concat([result, X], ignore_index = True)
    return result
```
