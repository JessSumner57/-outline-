```python
import pandas as pd
import re

import numpy as np



pd.set_option('display.max_rows', 300) # specifies number of rows to show
pd.options.display.float_format = '{:40,.4f}'.format # specifies default number format to 4 decimal places
```

```python
# read data from all sheets and store them in the same dataframe

def tenure_loop(file_name):
    '''
    This function reads data from each sheet and
    put them in the same dataframe.
    
    '''
    
    # excel sheet 2020 to 2012
    result = pd.DataFrame(columns = ['code', 'borough', 'own_outright', 'buying_with_mortgage', 'rented_from_local_authority_or_housing_association', 'rented_from_private_landowner', 'year'])
    for i in range(1, 16):
       
        X = pd.read_excel(file_name, sheet_name = i, usecols = 'A:F', skipfooter = 17, skiprows = 4,
                          header = None, names = ['code', 'borough', 'own_outright', 'buying_with_mortgage', 'rented_from_local_authority_or_housing_association', 'rented_from_private_landowner'])
        X['year'] = 2021 - int(f'{i}')
        
        result = pd.concat([result, X], ignore_index = True)
    return result
```

```python
file_name = 'tenure-population-borough.xlsx'
tenure = tenure_loop(file_name)
tenure.tail(n=10)
```
